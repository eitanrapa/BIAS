```python
from astropy.io import fits
import matplotlib.pyplot as plt

c = 299792458.0                                         # Speed of light - [c] = m/s
h_p = 6.626068e-34                                      # Planck's constant in SI units
k_b = 1.38065e-23                                       # Boltzmann constant in SI units
MJyperSrtoSI = 1e-20                                    # MegaJansky/Sr to SI units
GHztoHz = 1e9                                           # Gigahertz to hertz
HztoGHz = 1e-9                                          # Hertz to Gigahertz
TCMB = 2.725                                            # Canonical CMB in Kelvin
m = 9.109*10**(-31)                                     # Electron mass in kgs


#Galaxy cluster parameters
y_value = 6.0e-6
electron_temperature = 5.0 #KeV
peculiar_velocity = 0.000001 #km/s
betac=peculiar_velocity/3e5
time = 288000 #80 hours

def templates(freq, long, lang):
    
    #Read FITS file
    fname = '/data/bolocam/bolocam/erapaport/continuum.fits'
    hdu = fits.open(fname)
    image_data = hdu[0].data
    
    total_SIDES = np.zeros(751)
    #Rebinning for 3 arcmin resolution
    for col in range(6):
        for row in range(6):
            total_SIDES += image_data[:,long + row,lang + col]*MJyperSrtoSI
    total_SIDES = total_SIDES/36
    
    #Interpolate for specific frequencies
    sides_template = BIAS.interpolate(freq,total_SIDES,np.linspace(0,1500e9,751))
    
    #Galaxy cluster SZ template
    sz_template = BIAS.szpack_signal(freq, BIAS.y_to_tau(y_value,electron_temperature), electron_temperature, peculiar_velocity)
    
    template_total = [sz_template,sides_template]
    return template_total
```

```python
#Specify which SIDES pixel to look at 
long = 120
lat = 120

freq = np.linspace(80e9,720e9,1000)
templates_complete = templates(freq, long, lat)
  
plt.rc('xtick',labelsize=18)
plt.rc('ytick',labelsize=18)
plt.figure(figsize=(10,10))
  
#Plot SZ components
plt.plot(freq*HztoGHz,abs(templates_complete[0]),'--k',label='Total SZ, betac = 0',linewidth=2)
plt.plot(freq*HztoGHz,abs(BIAS.szpack_signal(freq,BIAS.y_to_tau(y_value,electron_temperature),electron_temperature,1e-11) - BIAS.classical_tsz(y_value,freq)),label='rSZ ' + str(electron_temperature) + ' keV')
#plt.plot(freq*HztoGHz,abs(templates_complete[0] - BIAS.szpack_signal(freq,BIAS.y_to_tau(y_value,electron_temperature),electron_temperature,1e-11)),\
#         label='kSZ ' + str(betac*3e5) + ' km/s')

plt.plot(freq*HztoGHz,abs(BIAS.classical_tsz(y_value,freq)),label='tSZ y=' + str(y_value))

#SIDES continuum model and emission line data
plt.plot(freq*HztoGHz,abs(templates_complete[1]),color='pink',label='SIDES continuum')
            
plt.xscale('log')
plt.yscale('log')
plt.xlabel('GHz',fontsize=20)
plt.ylabel('W/m^2/Hz/Sr',fontsize=20)

#Make xticks match as best as possible
plt.xticks(np.rint(np.logspace(np.log10(80), np.log10(7.2e2),num=9)),np.rint(np.logspace(np.log10(80), np.log10(7.2e2),num=9)))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', frameon=True, prop={'size':12}, ncol=1, title= '{} hour obs.'.format(time/3600))
plt.show()
```
